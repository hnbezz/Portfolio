# Car market insights -  a practical example

In this example, I work with the data of second hand cars for sales. This is real-life-like data, meaning that there are missing values, values that are incorrect and also with unnecessary features. I will use the data in order to create a linear regression model that would predict the price of second hand cars based on the information of this data.
In order to do that, we have to do different steps:

1) Explore the data

2) Clean the data

3) Relax Ordinary Least Square (OLS) assumptions

4) Work around non-linear features

5) Handle non-numerical (categorical) features

6) Create and test the model

Throughout this example I explain the reasoning behind the actions taken.


## Importing the relevant libraries

```python
# For this practical example I use the following libraries and modules
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

## Loading the raw data

```python
# Load the data from a .csv in the same folder
raw_data = pd.read_csv('Car_Market.csv')

# Let's explore the top 5 rows of the df
raw_data.head()
```

## Preprocessing

### Exploring the descriptive statistics of the variables

```python
# Here we check the description of all the data

"""
Some values are NaN due to the difference between numerical and categorical data. E.g., the model of a car are strings,
therefore it will not have a mean value.
"""
data = raw_data.copy()
raw_data.describe(include='all')
```

### Dealing with missing values

```python
"""
We have to check for missing values
"""
data.isnull().sum()
```

```python
"""
I simply drop the missing values, since they consist of less them 5% of the whole dataset. 
Then we check the data after dropping the missing values. We see from the count row that all features have the same amount
of data.
"""
data_no_mv = data.dropna(axis=0)
data_no_mv.describe(include='all')
```

### Exploring the PDFs

```python
"""
We can now explore the Probability Density Functions (PDFs) of the independent (inputs) and dependent (target) variables.
This will help us to observe outliers which can greatly reduce de predictive power of our model.
Moreover the PDF is often the basis which we can decide if we want to transform a feature or not.
"""

sns.distplot(data_no_mv['Price'])
```

From the graph above and the data description before, we can see that price has a mean value of ~20000, with a maximum value of 300000, which is 15 times the mean price value. Obviously there are outliers and it is important to remove them. This cannot be done blindly. We have to understand the logic behind it. It is very odd that a second hand car has a price of U$ 300,000. 

### Dealing with outliers

```python
"""
In this case I remove the top 1% of the data with the largest prices. 
"""
q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price']<q]
data_1.describe(include='all')
```

```python
"""Checking again the distribution to see if we impacted much on the distribution shape. We haven't, so we may continue."""
sns.distplot(data_1['Price'])
```

```python
"""Now we do the same thing for the other numerical data"""
sns.distplot(data_no_mv['Mileage'])
```

```python
"""Again we remove some outliers"""
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage']<q]
sns.distplot(data_2['Mileage'])
```

```python
"""
In the case of Engine Volume (named here EngineV) is rather strange. Analyzing the data we observe, and talking with
the hypothetical team of data acquisition, we observe that some of the missing values were actually entried with values
of 99.99 or 99. Moreover, there are wrong entries, such as 75.
"""
sns.distplot(data_no_mv['EngineV'])
```

```python
"""
Searching google, I see that most car engine volumes are below 6.5L. This emphasizes that we must be critical when analysing
and using data.

With that in mind, I remove all the data that has engine volume above 6.5 L.
"""

data_3 = data_2[data_2['EngineV']<6.5]
```

```python
# Following this graph, we realize we can actually treat EngineV as a categorical variable
# Even so, in this course we won't, but that's yet something else you may try on your own
sns.distplot(data_3['EngineV'])
```

```python
"""
The numerical data 'Year' is very similar with mileage and price, however for lower values. So I remove 
the first 1% of the data.
"""
sns.distplot(data_no_mv['Year'])
```

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]
sns.distplot(data_4['Year'])
```

```python
"""
When observations are removed, the original indexes were fixed throughout the data handling. It is a good practice
to reset the indexes.
"""
data_cleaned = data_4.reset_index(drop=True)
```

```python
data_cleaned.describe(include='all')
```

## Checking some the OLS assumptions

Now that we have the data cleaned of missed entries, wrong values and also of outliers, it is time to check the OLS assumptions. There are 6 of them:

1) Linearity between the dependent variables and independent variables

2) No endogeneity: basically the mean error must be zero. This can be checked after the linear regression is performed.

3) Normality: The errors follow a normal distribution.

4) Homoskedasticity: The variance of the errors is constant. A general fix is a log transformation of the sample.

5) No autocorrelation: for this we must use logic and analyze the source of the data.

6) Can we safely ignore multicollinearity? This can be checked through VIF (Variable inflation Factor).

```python
"""
We start with linearity. Here I plot the price as a function of Year, EngineV and Mileage. 
"""
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3)) #sharey -> share 'Price' as y
ax1.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax3.set_title('Price and Mileage')


plt.show()
```

```python
"""
From the subplots and the PDF of price, it is clear that therer is not linear relation, but rather exponention. In this case,
I use a log transformation of the price. 
"""
sns.distplot(data_cleaned['Price'])
```

### Relaxing the assumptions

```python
log_price = np.log(data_cleaned['Price'])
data_cleaned['log_price'] = log_price
data_cleaned.head()
```

```python
"""
Checking the scatter plots again to see if the transformation solved the issue.
"""

f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['log_price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['log_price'])
ax2.set_title('Log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['log_price'])
ax3.set_title('Log Price and Mileage')


plt.show()
```

```python
"""
I drop the older price feature, as i'll be using only the log_price
"""
data_cleaned = data_cleaned.drop(['Price'],axis=1)
```

### Creating dummies

Before checking for other OLS assumptions, we first have to preprocess the categorical data, so let's create dummies for them.

```python
"""Here we use the method get_dummies from pandas to create them. It is important to create drop the first for each feature, 
to avoid introducing multicollinearity.
"""

data_with_dummies = pd.get_dummies(data_cleaned, drop_first=True)
data_with_dummies.head()
```

### Multicollinearity

```python
"""sklearn does not have a built-in method to check multicollinearity. So I use variance_inflation_factor from
the statsmodels library.
"""
from statsmodels.stats.outliers_influence import variance_inflation_factor


"""To make this very simple I create a variable wich includes all features I need to check for multicollinearity."""

data_with_dummies = pd.get_dummies(data_cleaned, drop_first=True)
data_with_dummies.head()

variables = data_with_dummies.drop(['log_price'],axis=1)

vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
vif
```

```python
"""It is acceptable to remove the features which will lead to VIFs above 7. So I remove them"""
filterr = vif[vif['VIF']<7]
filterr = filterr.reset_index(drop=True)
variables_VIF = variables[variables.columns.intersection(filterr['Features'])]
variables_VIF
```

```python
"""Now we check again VIF"""
variables2 = variables_VIF

vif2 = pd.DataFrame()
vif2["VIF"] = [variance_inflation_factor(variables2.values, i) for i in range(variables2.shape[1])]
vif2["Features"] = variables2.columns
vif2
```

```python
"""It seems that Year and registration lead to collinearity again, so we again remove them and check VIF"""
variables = variables2.drop(['Year','Registration_yes'],axis=1)
vif3 = pd.DataFrame()
vif3["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif3["Features"] = variables.columns
vif3
```

```python
"""Now it seems that VIF is acceptable, let's check how the variables are"""

variables.describe()
```

## Linear regression model

### Declare the inputs and the targets

```python
# The target(s) (dependent variable) is 'log price'
targets = data_with_dummies['log_price']

# The inputs are everything BUT the dependent variable, so we can simply drop it
inputs = variables
```

### Scale the data

```python
# Import the scaling module
from sklearn.preprocessing import StandardScaler

# Create a scaler object
scaler = StandardScaler()
# Fit the inputs (calculate the mean and standard deviation feature-wise)
scaler.fit(inputs)
```

```python
# Scale the features and store them in a new variable (the actual scaling procedure)
inputs_scaled = scaler.transform(inputs)
```

### Train Test Split

```python
# Import the module for the split
from sklearn.model_selection import train_test_split

# Split the variables with an 80-20 split and some random state
# To have the same split as mine, use random_state = 365
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
# Create a linear regression object
reg = LinearRegression()
# Fit the regression with the scaled TRAIN inputs and targets
reg.fit(x_train,y_train)
```

```python
# Let's check the outputs of the regression
# I'll store them in y_hat as this is the 'theoretical' name of the predictions
y_hat = reg.predict(x_train)
```

```python
# The simplest way to compare the targets (y_train) and the predictions (y_hat) is to plot them on a scatter plot
# The closer the points to the 45-degree line, the better the prediction
plt.scatter(y_train, y_hat)
# Let's also name the axes
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
# Sometimes the plot will have different scales of the x-axis and the y-axis
# This is an issue as we won't be able to interpret the '45-degree line'
# We want the x-axis and the y-axis to be the same
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# Another useful check of our model is a residual plot
# We can plot the PDF of the residuals and check for anomalies
sns.distplot(y_train - y_hat)

# Include a title
plt.title("Residuals PDF", size=18)

# In the best case scenario this plot should be normally distributed
# In our case we notice that there are many negative residuals (far away from the mean)
# Given the definition of the residuals (y_train - y_hat), negative values imply
# that y_hat (predictions) are much higher than y_train (the targets)
# This is food for thought to improve our model
```

```python
# Find the R-squared of the model
reg.score(x_train,y_train)

# Note that this is NOT the adjusted R-squared
# in other words... find the Adjusted R-squared to have the appropriate measure :)
```

### Finding the weights and bias

```python
# Obtain the bias (intercept) of the regression
reg.intercept_
```

```python
# Obtain the weights (coefficients) of the regression
reg.coef_

# Note that they are barely interpretable if at all
```

```python
# Create a regression summary where we can compare them with one-another
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
# Check the different categories in the 'Brand' variable
data_cleaned['Brand'].unique()

# In this way we can see which 'Brand' is actually the benchmark
```

## Testing

```python
# Once we have trained and fine-tuned our model, we can proceed to testing it
# Testing is done on a dataset that the algorithm has never seen
# Luckily we have prepared such a dataset
# Our test inputs are 'x_test', while the outputs: 'y_test' 
# We SHOULD NOT TRAIN THE MODEL ON THEM, we just feed them and find the predictions
# If the predictions are far off, we will know that our model overfitted
y_hat_test = reg.predict(x_test)
```

```python
# Create a scatter plot with the test targets and the test predictions
# You can include the argument 'alpha' which will introduce opacity to the graph
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.plot([7,12],[7,12],color='red')
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# Finally, let's manually check these predictions
# To obtain the actual prices, we take the exponential of the log_price
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
# We can also include the test targets in that data frame (so we can manually compare them)
df_pf['Target'] = np.exp(y_test)
df_pf

# Note that we have a lot of missing values
# There is no reason to have ANY missing values, though
# This suggests that something is wrong with the data frame / indexing
```

```python
# After displaying y_test, we find what the issue is
# The old indexes are preserved (recall earlier in that code we made a note on that)
# The code was: data_cleaned = data_4.reset_index(drop=True)

# Therefore, to get a proper result, we must reset the index and drop the old indexing
y_test = y_test.reset_index(drop=True)

# Check the result
y_test.head()
```

```python
# Let's overwrite the 'Target' column with the appropriate values
# Again, we need the exponential of the test log price
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
# Additionally, we can calculate the difference between the targets and the predictions
# Note that this is actually the residual (we already plotted the residuals)
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']

# Since OLS is basically an algorithm which minimizes the total sum of squared errors (residuals),
# this comparison makes a lot of sense
```

```python
# Finally, it makes sense to see how far off we are from the result percentage-wise
# Here, we take the absolute difference in %, so we can easily order the data frame
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```

```python
# Exploring the descriptives here gives us additional insights
df_pf.describe()
```

```python
# Sometimes it is useful to check these outputs manually
# To see all rows, we use the relevant pandas syntax
pd.options.display.max_rows = 999
# Moreover, to make the dataset clear, we can display the result with only 2 digits after the dot 
pd.set_option('display.float_format', lambda x: '%.2f' % x)
# Finally, we sort by difference in % and manually check the model
df_pf.sort_values(by=['Difference%'])
```
